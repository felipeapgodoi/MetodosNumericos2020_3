```
# Universidade Federal de Uberlândia
# Programa de Pós Graduação Em Engenharia Química
# Métodos Numéricos Aplicados à Engenharia Química
# Aluno: Felipe Augusto Paes de Godoi
# Atividade 03: Questão 2-Calcule as concentrações de equilíbrio Ca,Cb,Cc,Cd,Cx,Cy,Cz
# Sistema gasoso em reator batelada de volume constante
# Reações:
# A + B <-> C + D
# B + C <-> X + Y
# A + X <-> Z
# Constantes de equilíbrio:
# Kc1=Cc*Cd/(Ca*Cb)
# Kc2=Cx*Cy/(Cb*Cc)
# Kc3=Cz/(Ca*Cx)
# Concentrações:
# Ca=Ca0-Cd-Cz
# Cb=Cb0-Cd-Cy
# Cc=Cd-Cy
# Cy=Cx+Cz
# Sistema de equações resultantes, em termos de Cd,Cy,Cz:
# (Kc1-1)*Cd**2+(Kc1+1)*Cd*Cy+Kc1*Cd*Cz+(-Kc1*Ca0-Kc1*Cb0)*Cd-Kc1*Ca0*Cy+Kc1*Cy*Cz-Kc1*Cb0*Cz+Kc1*Ca0*Cb0=0
# -Kc2*Cd**2+Kc2*Cb0*Cd+(Kc2-1)*Cy**2-Kc2*Cb0*Cy+Cy*Cz=0
# -Kc3*Cd*Cy+Kc3*Cd*Cz+Kc3*Ca0*Cy-Kc3*Cy*Cz+(-Kc3*Ca0-1)*Cz+Kc3*Cz**2=0

import numpy  as np
import pandas as pd

t=1e-5 # t is the learing rate
Ite=100000 # number of iteractions
x0=np.array([0.5,0.5,0.5]) # initial guess of x

def Functions(x):

  Kc1=1.06
  Kc2=2.63
  Kc3=5
  Ca0=1.5
  Cb0=1.5
  F=np.zeros(3)
  F[0]=(Kc1-1)*x[0]**2+(Kc1+1)*x[0]*x[1]+Kc1*x[0]*x[2]+(-Kc1*Ca0-Kc1*Cb0)*x[0]-Kc1*Ca0*x[1]+Kc1*x[1]*x[2]-Kc1*Cb0*x[2]+Kc1*Ca0*Cb0
  F[1]=-Kc2*x[0]**2+Kc2*Cb0*x[0]+(Kc2-1)*x[1]**2-Kc2*Cb0*x[1]+x[1]*x[2]
  F[2]=-Kc3*x[0]*x[1]+Kc3*x[0]*x[2]+Kc3*Ca0*x[1]-Kc3*x[1]*x[2]+(-Kc3*Ca0-1)*x[2]+Kc3*x[2]**2
  return F

def fun_and_grad(x):

  dim=len(x)
  eps=1e-5

  F=Functions(x)

  up=x.copy()
  low=x.copy()
  Z=0

  for i in range(dim):
    Z=Z+abs(F[i])
  
  
  gradZ=np.zeros(3)
  
  for i in range(dim):
    
    up=x.copy()
    low=x.copy()
    up[i]=x[i]+eps
    low[i]=x[i]-eps
    Zup=0
    Zlow=0
    Flow=Functions(low)
    Fup=Functions(up)

    for k in range(dim):
      Zup=Zup+abs(Fup[k])
      Zlow=Zlow+abs(Flow[k])
  
    gradZ[i]=(Zup-Zlow)/(2*eps)

  return Z,gradZ

dim=len(x0)
x=np.copy(x0)
print("x=",x)
lista1=[]

for i in range(Ite):
  Z,gradZ=fun_and_grad(x)
  #print("x=",x)
  #print("gradZ=",gradZ)
  x=x-gradZ*t

  if((i+1)%2000==0):
    print("i=",i+1,"x=",x,"gradZ=",gradZ," z=%.5e"%(Z))
  
```
