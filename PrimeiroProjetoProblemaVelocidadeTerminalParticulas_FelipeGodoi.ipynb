```
# Universidade Federal de Uberlândia
# Programa de Pós Graduação Em Engenharia Química
# Métodos Numéricos Aplicados à Engenharia Química
# Aluno: Felipe Augusto Paes de Godoi
# Primeiro Projeto: Problema 5- Velocidade terminal de sedimentação de partículas
# Questão 1- calcule a velocidade terminal de partículas de carvão em água a 298.15 K:
# Velocidade terminal das parículas:
# vt=((4*g*(rhop-rho)*Dp)/(3*CD*rho))**0.5
# Coeficiente de arrasto:
# a) CD=24/Re para Re < 0,1
# b) CD=24*(1+0.14*Re**0.7)/Re para 0.1 <= Re <= 1000
# c) CD=0.44 para 1000 < Re < 350000
# d) CD=0.19-(8*10**4)/Re para 350000 < Re
# Número de Reynolds:
# Re=(Dp*vt*rho)/mi
# rhop: densidade das partículas
# rho: densidade do fluido
# g: aceleração da gravidade
# Dp: diâmetro das partículas
# mi: viscosidade do fluido
# Equações para o cálculo de vt para cada expressão de CD, substituindo Re na expressão de vt:
# a) CD=24/Re para Re < 0.1:
# vt=(4*g*(rhop-rho)*Dp**2.0)/(3*24*mi)
# b) CD=24*(1+0.14*Re**0.7)/Re para 0.1 <= Re <= 1000
# 3*24*0.14*Dp**0.7*rho**0.7*mi**0.3*vt**1.7+3*24*mi*vt-4*g*(rhop-rho)*Dp**2.0=0
# c) CD=0.44 para 1000 < Re <= 350000
# vt=((4*g*(rhop-rho)*Dp)/(3*0.44*rho))**0.5
# d) CD=0.19-(8*10**4)/Re para 350000 < Re
# 3*0.19*Dp*rho*vt**2-3*8*10**4*mi*vt-4*g*(rhop-rho)*Dp**2.0=0

# a) Supondo que Re é menor que 0.1

rhop=1800.0
rho=994.6
mi=8.931*10**-4
Dp=0.208*10**-3
g=9.80665

vt=(4*g*(rhop-rho)*Dp**2.0)/(3*24*mi)
Re=(Dp*vt*rho)/mi

print('a) velocidade terminal (m/s)=', vt)
print('número de Reynolds calculado=', Re)

if Re < 0.1:
  print('O número de Reynolds calculado se encontra na faixa de aplicação da respectiva expressão de Coeficiente de arrasto')
else:
  print('O número de Reynolds calculado não se encontra na faixa de aplicação da respectiva expressão de Coeficiente de arrasto')

# b) Supondo que Re se encontra no intervalo entre 0.1 e 1000

def bisection(f, x_L, x_R, eps,num):
    f_L = f(x_L)
    if f_L*f(x_R) > 0:
        print( "Error! Function does not have opposite signs at interval endpoints!")
        sys.exit(1)
    x_M = float(x_L + x_R)/2.0
    f_M = f(x_M)
    iteration_counter = 1

    while abs(f_M) > eps:
        if f_L*f_M > 0:   # i.e. same sign
            x_L = x_M
            f_L = f_M
        else:
            x_R = x_M
        x_M = float(x_L + x_R)/2
        f_M = f(x_M)
        iteration_counter += 1

        if(iteration_counter>num):
          break

    return x_M, iteration_counter

def f(x):
  
  rhop=1800.0
  rho=994.6
  mi=8.931*10**-4
  Dp=0.208*10**-3
  g=9.80665

  return 3*24*0.14*Dp**0.7*rho**0.7*mi**0.3*x**1.7+3*24*mi*x-4*g*(rhop-rho)*Dp**2.0

xl = (0.2*mi)/(Dp*rho);   xd = (990*mi)/(Dp*rho); total=200;eps=1.0e-6

vt, no_iterations = bisection(f, xl, xd ,eps ,total)
print('b) velocidade terminal (m/s)=', vt, 'no_iterations=',no_iterations)

Re=(Dp*vt*rho)/mi

print('número de Reynolds calculado=', Re)

if 0.1 <= Re <= 1000:
  print('O número de Reynolds calculado se encontra na faixa de aplicação da respectiva expressão de Coeficiente de arrasto')
else:
  print('O número de Reynolds calculado não se encontra na faixa de aplicação da respectiva expressão de Coeficiente de arrasto')

# c) Supondo que Re se encontra no intervalo entre 1000 e 350000

rhop=1800.0
rho=994.6
mi=8.931*10**-4
Dp=0.208*10**-3
g=9.80665

vt=((4*g*(rhop-rho)*Dp)/(3*0.44*rho))**0.5
Re=(Dp*vt*rho)/mi

print('c) velocidade terminal (m/s)=', vt)
print('número de Reynolds calculado=', Re)

if 1000 < Re <= 350000:
  print('O número de Reynolds calculado se encontra na faixa de aplicação da respectiva expressão de Coeficiente de arrasto')
else:
  print('O número de Reynolds calculado não se encontra na faixa de aplicação da respectiva expressão de Coeficiente de arrasto')

# d) Supondo que Re é maior que 350000

def bisection(f, x_L, x_R, eps,num):
    f_L = f(x_L)
    if f_L*f(x_R) > 0:
        print( "Error! Function does not have opposite signs at interval endpoints!")
        sys.exit(1)
    x_M = float(x_L + x_R)/2.0
    f_M = f(x_M)
    iteration_counter = 1

    while abs(f_M) > eps:
        if f_L*f_M > 0:   # i.e. same sign
            x_L = x_M
            f_L = f_M
        else:
            x_R = x_M
        x_M = float(x_L + x_R)/2
        f_M = f(x_M)
        iteration_counter += 1

        if(iteration_counter>num):
          break

    return x_M, iteration_counter

def f(x):
  
  rhop=1800.0
  rho=994.6
  mi=8.931*10**-4
  Dp=0.208*10**-3
  g=9.80665

  return 3*0.19*Dp*rho*x**2-3*8*10**4*mi*x-4*g*(rhop-rho)*Dp**2.0

xl = (360000*mi)/(Dp*rho);   xd = (10**8*mi)/(Dp*rho); total=200;eps=1.0e-6

vt, no_iterations = bisection(f, xl, xd ,eps ,total)
print('d) velocidade terminal (m/s)=', vt, 'no_iterations=',no_iterations)

Re=(Dp*vt*rho)/mi

print('número de Reynolds calculado=', Re)

if Re > 350000:
  print('O número de Reynolds calculado se encontra na faixa de aplicação da respectiva expressão de Coeficiente de arrasto')
else:
  print('O número de Reynolds calculado não se encontra na faixa de aplicação da respectiva expressão de Coeficiente de arrasto')
```
